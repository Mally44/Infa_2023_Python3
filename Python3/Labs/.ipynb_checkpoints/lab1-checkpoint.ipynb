```python
#квадрат
import turtle 
turtle.shape ('turtle')
turtle.color ('blue')

a = [(90,100),(90,100),(90,100),(90,100)]

for t in a:
    angle, lengs = t
    turtle.forward(lengs)
    turtle.left(angle)
```

```python
#паук
import turtle 
turtle.shape ('turtle')
turtle.width(2)
turtle.speed(2)

for t in range (12):
    lengs = 100
    angle = 30
    turtle.right(angle)
    turtle.forward(lengs)
    turtle.stamp()
    turtle.backward(lengs)
    
turtle.hideturtle()
```

```python
#квадратная спираль
import turtle 
turtle.shape ('turtle')

for t in range (60):
    lenght = t * 2
    turtle.right(90)
    turtle.forward(lenght)
    
```

```python
#многоугольники
import turtle

delta = 10 # изменеие радиуса
n = 3 # число сторон первого многоугольника 

for A in range(10):
    turtle.penup()
    turtle.setheading(90)
    turtle.goto((A+1)*delta, 0)
    turtle.pendown()
    dN = A + n
    radius = (A+1) * delta
    turtle.circle(radius, 360, dN)
```

```python
#смайлик
import turtle
turtle.shape('turtle')
turtle.speed(10)

def lico():
    turtle.color('black', 'yellow')
    turtle.begin_fill()
    turtle.circle(120)
    turtle.end_fill()
    
def glaz():
    turtle.color('blue')
    turtle.begin_fill()
    turtle.circle(12,360)
    turtle.end_fill()

def nose():
    turtle.width(10)
    turtle.color('black')
    turtle.setheading(270)
    turtle.fd(40)
    
def smile():
    turtle.width(11)
    turtle.color('red')
    turtle.rt(180)
    turtle.circle(60,-180)
    
lico()
turtle.up()
turtle.goto(-40,165)
turtle.down()
glaz()
turtle.up()
turtle.goto(40,165)
turtle.down()
glaz()
turtle.up()
turtle.goto(0,150)
turtle.down()
nose()
turtle.up()
turtle.goto(60,80)
turtle.down()
smile()
turtle.ht()
```

```python
# зведы
import turtle
turtle.shape('turtle')

def stars(n):
    for t in range (n):
        angle = 180 -(180 / n)
        turtle.lt(angle)
        turtle.fd(150)
        
turtle.penup()        
turtle.goto(-200,0)
turtle.down()
turtle.setheading(180)
stars(5)
turtle.penup()        
turtle.goto(100,0)
turtle.down()
stars(11)
turtle.ht()

    
```
