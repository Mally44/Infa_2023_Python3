```python
import pygame
from pygame.draw import *
from random import randint

class Square():
    def __init__(self):
        self.x = randint(100, 700)
        self.y = randint(100, 500)
        self.width = randint(50, 80)
        self.height = self.width
        self.point = 0
        self.color = COLORS[randint(0, 5)]
        self.speedx = randint(-10, 10)
        self.speedy = randint(-10, 10)

    def create(self):
        '''
        Создает квадраты разных цветов и размеров
        '''
        self.move()
        rect(screen, self.color, (self.x, self.y, self.width, self.height))

    def move(self):
        '''
        Двигает квадраты и отражает от стен
        '''
        self.x += self.speedx
        self.y += self.speedy

        if self.x >= W - self.width or self.x <= 0:
            self.speedx = -self.speedx
        if self.y >= H - self.height or self.y <= 0:
            self.speedy = -self.speedy

    def click(self, event):
        '''
        Функция, которая прибавляет одно очко при нажатие квадрат,
        а также при нажатии на мышку меняет положение квадрата, его размер,
        цвет и скорость передвижения
        '''

        if (event.pos[0] >= self.x and event.pos[0] <= self.x + self.width
            ) and (event.pos[1] >= self.y
                   and event.pos[1] <= self.y + self.height):
            self.point += 1

        self.x = randint(100, 700)
        self.y = randint(100, 500)
        self.width = randint(50, 80)
        self.height = self.width
        self.color = COLORS[randint(0, 5)]
        self.speedx = randint(-10, 10)
        self.speedy = randint(-10, 10)

    def Points(self):
        '''
        Выводить на экран количество очков квадратов в игре
        '''
        f = pygame.font.Font(None, 36)
        text = f.render(("Points_square:" + str(self.point)), 1, (31, 10, 214))
        text_rect = text.get_rect(topleft=(0, 0))
        screen.blit(text, text_rect)
```
