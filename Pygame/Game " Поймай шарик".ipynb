```python
import pygame
from pygame.draw import *
from random import randint

pygame.init()

W, H = 1100, 700
FPS = 30
screen = pygame.display.set_mode((W, H))



RED = (255, 0, 0)
BLUE = (0, 0, 255)
YELLOW = (255, 255, 0)
GREEN = (0, 255, 0)
MAGENTA = (255, 0, 255)
CYAN = (0, 255, 255)
BLACK = (0, 0, 0)
COLORS = [RED, BLUE, YELLOW, GREEN, MAGENTA, CYAN]


class Ball():
    def __init__(self):
        self.x = randint(100, 600)
        self.y = randint(100, 500)
        self.r = randint(30, 50)
        self.point = 0
        self.color = COLORS[randint(0, 5)]
        self.speedx = randint(-10, 10)
        self.speedy = randint(-10, 10)

    def create(self):
        '''
        Создает шарики разных цветов и размеров
        '''
        self.move()
        circle(screen, self.color, (self.x, self.y), self.r)

    def move(self):
        '''
        Двигает шарики и отражает от стен
        '''
        self.x += self.speedx
        self.y += self.speedy

        if self.x >= W - self.r or self.x <= self.r:
            self.speedx = -self.speedx
        if self.y >= H - self.r or self.y <= self.r:
            self.speedy = -self.speedy

    def click(self, event):
        '''
        Функция, которая прибавляет одно очко при нажатие на шар,
        а также при нажатии на мышку меняет положение шарика, его размер,
        цвет и скорость передвижения
        '''
        if (event.pos[0] - self.x)**2 + (event.pos[1] - self.y)**2 <= self.r**2:
            self.point += 1

        self.x = randint(100, 700)
        self.y = randint(100, 500)
        self.r = randint(30, 50)
        self.color = COLORS[randint(0, 5)]
        self.speedx = randint(-10, 10)
        self.speedy = randint(-10, 10)

    def Points(self):
        '''
        Выводить на экран количество очков шариков в игре
        '''
        f = pygame.font.Font(None, 36)
        text = f.render(("Points_ball:" + str(self.point)), 1, (255, 0, 0))
        text_rect = text.get_rect(topleft= (0, 30))
        screen.blit(text, text_rect)


class Square():
    def __init__(self):
        self.x = randint(100, 700)
        self.y = randint(100, 500)
        self.width = randint(50, 80)
        self.height = self.width
        self.point = 0
        self.color = COLORS[randint(0, 5)]
        self.speedx = randint(-10, 10)
        self.speedy = randint(-10, 10)

    def create(self):
        '''
        Создает квадраты разных цветов и размеров
        '''
        self.move()
        rect(screen, self.color, (self.x, self.y, self.width, self.height))

    def move(self):
        '''
        Двигает квадраты и отражает от стен
        '''
        self.x += self.speedx
        self.y += self.speedy

        if self.x >= W - self.width or self.x <= 0:
            self.speedx = -self.speedx
        if self.y >= H - self.height or self.y <= 0:
            self.speedy = -self.speedy

    def click(self, event):
        '''
        Функция, которая прибавляет одно очко при нажатие квадрат,
        а также при нажатии на мышку меняет положение квадрата, его размер,
        цвет и скорость передвижения
        '''

        if (event.pos[0] >= self.x and event.pos[0] <= self.x + self.width
            ) and (event.pos[1] >= self.y
                   and event.pos[1] <= self.y + self.height):
            self.point += 1
        self.x = randint(100, 700)
        self.y = randint(100, 500)
        self.width = randint(50, 80)
        self.height = self.width
        self.color = COLORS[randint(0, 5)]
        self.speedx = randint(-10, 10)
        self.speedy = randint(-10, 10)

    def Points(self):
        '''
        Выводить на экран количество очков квадратов в игре
        '''
        f = pygame.font.Font(None, 36)
        text = f.render(("Points_square:" + str(self.point)), 1, (31, 10, 214))
        text_rect = text.get_rect(topleft = (0,0))
        screen.blit(text, text_rect)


a, b = map(int, input().split()) # how many balls and squares
balls, squares = [Ball() for i in range(a)], [Square() for i in range(b)]
ball = Ball()
square = Square()


pygame.display.update()
clock = pygame.time.Clock()
finished = False

while not finished:
    clock.tick(FPS)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            finished = True
        elif event.type == pygame.MOUSEBUTTONDOWN:
            for ball in balls:
                ball.click(event)
            for square in squares:
                square.click(event)

    screen.fill(BLACK)
    for ball in balls:
        ball.create()
    for square in squares:
        square.create()
    ball.Points()
    square.Points()
    pygame.display.update()

pygame.display.update()
pygame.quit()
```
